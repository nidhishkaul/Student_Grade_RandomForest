```python
## Importing necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Reading the data
data = pd.read_csv('student_performance_large_dataset.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
## Checking for nulls
data.isnull().sum()
```

```python
## Checking for duplicates
data.duplicated().sum()
```

```python
## Dropping unnecessary columns
data.drop('Student_ID',axis=1,inplace=True)
```

```python
## Handling categorical variables using OHE
categorical_cols = [i for i in data.columns if data[i].dtype == 'object']
```

```python
categorical_cols.remove('Final_Grade')
```

```python
categorical_cols
```

```python
data = pd.get_dummies(data,columns=categorical_cols,drop_first=True)
```

```python
## Applying ordinal mapping on the final grade
grades = {'A':0,'B':1,'C':2,'D':3}
data['Final_Grade'] = data['Final_Grade'].map(grades)
```

```python
data.head()
```

```python
## Independent and Dependent Features
x = data.drop('Final_Grade',axis=1)
y = data['Final_Grade']
```

```python
## Train-Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
## Standard Scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
## Model Training
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
```

```python
model.fit(x_train_scaled,y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
## Performance Metrics
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
